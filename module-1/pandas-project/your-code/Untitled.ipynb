```python
null_cols = sharks.isnull().sum()
null_cols[null_cols > 0]
```

```python
drop_cols = list(null_cols[null_cols > 5000].index)
sharks= sharks.drop(drop_cols, axis=1)
```

```python
sharks.head()
```

```python
low_variance = []

for col in sharks._get_numeric_data():
    minimum = min(sharks[col])
    ninety_perc = np.percentile(sharks[col], 90)
    if ninety_perc == minimum:
        low_variance.append(col)
print(low_variance)
```
