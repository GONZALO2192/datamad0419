```python
bins = [0,2,6,10,19]
edades = ['0-1', '2-5', '6-10', '10-19']
ani['bins'] = pd.cut(ani['age_upon_outcome_years'], bins ,labels=edades)
   
ani.head()



```

```python

w=ani.groupby('bins')['outcome_type'].value_counts()
w_1=w.plot(kind='bar', color=['g', 'b', 'r','y','purple','orange','pink'])
```

```python
plt.savefig('w_1')
plt.show()
```
